```python
#https://stackabuse.com/implementing-pca-in-python-with-scikit-learn/

#importing the libraries required for reading the file 
import numpy as np  
import pandas as pd 
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"  
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']  
dataset = pd.read_csv(url, names=names) 
```

```python
dataset.head()
```

```python
X = dataset.drop('Class', 1)  
y = dataset['Class'] 
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  
```

```python
#PCA performs best with a normalized feature set. We will perform standard scalar normalization to normalize our feature set
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()  
X_train = sc.fit_transform(X_train)  
X_test = sc.transform(X_test) 


```

```python
print('#####################################################################################')
print(X_train)
print('#####################################################################################')
print(X_test)
print('#####################################################################################')
print(y_train)
print('#####################################################################################')
print(y_test)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#print(X_test)
X_test.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA()  
X_train = pca.fit_transform(X_train)  
X_test = pca.transform(X_test)  
```

```python
explained_variance = pca.explained_variance_ratio_  
```

```python
print(explained_variance)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)  
X_train = pca.fit_transform(X_train)  
X_test = pca.transform(X_test) 
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=2, random_state=0)  
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)  
```

```python
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score

new = str(accuracy_score(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)  
print(cm)  
print('Accuracy' + new)
```
